```python
import csv
with open('height-weight.csv') as f:
  reader = csv.reader(f)
  data = list(reader)
print(data)
```

```python
data.pop(0)
print(data)
```

```python
newData=[]
for i in range(len(data)):
  newData.append(float(data[i][2]))
print(newData)
```

```python
totalSum=0
for i in newData:
  totalSum = totalSum + i
print(totalSum)
```

```python
mean = totalSum/len(newData)
print(mean)
print(len(newData))
```

```python
newData.sort()
print(newData)
```

```python
totalNumber = len(newData)
if totalNumber % 2 ==0:
  median1=newData[totalNumber//2]
  median2=newData[totalNumber//2]
  median = (median1 + median2)/2
else:
  median = newData[totalNumber//2]

print(median)
```

```python
from collections import Counter
something = Counter(newData)
print(something)
mode = something.most_common(1)
print(mode)
```
