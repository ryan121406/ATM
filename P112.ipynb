```python
q1 = df["quant_saved"].quantile(0.25) 
q3 = df["quant_saved"].quantile(0.75) 
iqr = q3-q1 
lower_whisker = q1 - 1.5*iqr
upper_whisker = q3 + 1.5*iqr
#Creating a new DataFrame 
new_df = df[df["quant_saved"] < upper_whisker]
new_df
```

```python
fig = px.scatter(new_df, y="quant_saved",color='wealthy')
fig.show()
```

```python
fig = ff.create_distplot([new_df["quant_saved"]], ["Savings"], show_hist=False) 
fig.show()
```

```python
savedMean = new_df['quant_saved'].mean()
savedMedian = new_df['quant_saved'].median()
savedMode = new_df['quant_saved'].mode()
savedStd = new_df['quant_saved'].std()
print(savedMean)
print(savedMedian)
print(savedMode)
print(savedStd)
```

```python
myList = new_df['quant_saved'].tolist()

def findSampleMean():
  sample = []
  for i in range(0,100):
    randomIndex = rd.randint(0,len(myList)-1)
    value = myList[randomIndex]
    sample.append(value)
  meanSample = st.mean(sample)
  return meanSample

meanList = []

for i in range(0,1000):
  f = findSampleMean()
  meanList.append(f)

print(meanList)

meanListMean = st.mean(meanList)
print(meanListMean)
meanListstdev = st.stdev(meanList)
print(meanListstdev)
```

```python
fig = ff.create_distplot([meanList], ["Savings"], show_hist=False) 
fig.show()
```
