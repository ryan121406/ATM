```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import pandas as pd
import statistics as st
import random as rd

df = pd.read_csv("medium_data.csv")
df
```

```python
myData = df['reading_time'].tolist()
myData
```

```python
myMean = st.mean(myData)
print(myMean)
myStd = st.stdev(myData)
print(myStd)
```

```python
fig = ff.create_distplot([myData],['Reading Time Data'])
fig.add_trace(go.Scatter(x=[myMean, myMean], y=[0,0.2], mode = "lines", name = "Mean"))
```

```python
sampleData = []

for i in range(1,100):
  randomIndex = rd.randint(0,len(myData))
  sampleValue = myData[randomIndex]
  sampleData.append(sampleValue)

print(sampleData)
print(myData)
```

```python
def listSampleDataMean():
  for i in range (0,30):
    randomIndex = rd.randint(0, len(myData)-1)      
    mySampleValue = myData[randomIndex]           
    sampleData.append(mySampleValue)
  mySampleMean = st.mean(sampleData)   
  return mySampleMean
```

```python
meanList = []

for i in range(0,100):
  temp = listSampleDataMean()
  meanList.append(temp)

print(meanList)

myMeanList = st.mean(meanList)
myMeanListStd = st.stdev(meanList)
```

```python
fig1 = ff.create_distplot([meanList],['Mean List'])
#fig1.add_trace(go.Scatter(x=[myMeanList, myMeanList], y=[0,0.2], mode = "lines", name = "Mean"))
fig1.show()
```

```python
firstStd_start, firstStd_end, secondStd_start, secondStd_end, thirdStd_start, thirdStd_end = myMeanList-myMeanListStd, myMeanList+myMeanListStd, myMeanList-(2*myMeanListStd), myMeanList+(2*myMeanListStd), myMeanList-(3*myMeanListStd), myMeanList+(3*myMeanListStd) 
```

```python
fig0 = ff.create_distplot([meanList], ["student marks"], show_hist=False)
fig0.add_trace(go.Scatter(x=[myMeanList, myMeanList], y=[0, 10], line = dict(color='green', width=4,dash='dash'), name="MEAN"))
fig0.add_trace(go.Scatter(x=[firstStd_start, firstStd_start], y=[0, 10], line = dict(color='red', width=2), name="STANDARD DEVIATION 1 START"))
fig0.add_trace(go.Scatter(x=[firstStd_end, firstStd_end], y=[0, 10], line = dict(color='red', width=2), name="STANDARD DEVIATION 1 END"))
fig0.add_trace(go.Scatter(x=[secondStd_start, secondStd_start], y=[0, 10], line = dict(color='blue', width=2), name="STANDARD DEVIATION 2 START"))
fig0.add_trace(go.Scatter(x=[secondStd_end, secondStd_end], y=[0, 10], line = dict(color='blue', width=2), name="STANDARD DEVIATION 2 END"))
fig0.add_trace(go.Scatter(x=[thirdStd_start,thirdStd_start], y=[0,10], line = dict(color='orange', width=2), name="STANDARD DEVIATION 3 START"))
fig0.add_trace(go.Scatter(x=[thirdStd_end,thirdStd_end], y=[0,10],line = dict(color='orange', width=2), name="STANDARD DEVIATION 3 END"))
fig0.show()
```

```python
myNewSampleMean = listSampleDataMean()
print(myNewSampleMean)
```

```python
fig3 = ff.create_distplot([meanList], ["student marks"], show_hist=False)
fig3.add_trace(go.Scatter(x=[myMeanList, myMeanList], y=[0, 10], line = dict(color='green', width=4,dash='dash'), name="MEAN"))
fig3.add_trace(go.Scatter(x=[firstStd_start, firstStd_start], y=[0, 10], line = dict(color='red', width=2), name="STANDARD DEVIATION 1 START"))
fig3.add_trace(go.Scatter(x=[firstStd_end, firstStd_end], y=[0, 10], line = dict(color='red', width=2), name="STANDARD DEVIATION 1 END"))
fig3.add_trace(go.Scatter(x=[secondStd_start, secondStd_start], y=[0, 10], line = dict(color='blue', width=2), name="STANDARD DEVIATION 2 START"))
fig3.add_trace(go.Scatter(x=[secondStd_end, secondStd_end], y=[0,10], line = dict(color='blue', width=2), name="STANDARD DEVIATION 2 END"))
fig3.add_trace(go.Scatter(x=[thirdStd_start,thirdStd_start], y=[0,10], line = dict(color='orange', width=2), name="STANDARD DEVIATION 3 START"))
fig3.add_trace(go.Scatter(x=[thirdStd_end,thirdStd_end], y=[0,10],line = dict(color='orange', width=2), name="STANDARD DEVIATION 3 END"))
fig3.add_trace(go.Scatter(x=[myNewSampleMean, myNewSampleMean], y=[0,10], line = dict(color='black', width=2,dash='dash'), name="MEAN OF SAMPLE"))


fig3.show()
```
