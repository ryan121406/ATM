```python
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go
import statistics as st
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/whitehatjr/datasets/master/main.csv')
df
```

```python
score = df['TOEFL Score'].tolist()
admissionChance = df['Chance of Admit '].tolist()
```

```python
fig = px.scatter(x = score, y = admissionChance)
fig.show()
```

```python
m =0.018599296811431736 
c = -1.273400470515882
y = []
```

```python
for x in score:
  value = m*x + c
  y.append(value)
  
print(y)
```

```python
fig.update_layout(shapes =[dict(type = 'line',y0 = min(y),y1=max(y),x0 = min(score),x1 = max(score))])
fig.show()
```

```python
scoreArray = np.array(score)
chanceArray = np.array(admissionChance)
```

```python
m,c = np.polyfit(scoreArray,chanceArray,1)
print(m,c)

x = 70
y=m*x + c
print(y)
```
